# 4. Cleaning the Dataset

# a. Removing the redundunt rows of data

```python
# Checking redundunt rows of data
# Dropping columns which are not required for further analysis
raw_df.drop(['Zomato URL', 'Address'], axis = 1, inplace=True) 
```

# b. Removing the Null Records

```python
# Checking for Null records
raw_df.isnull().sum()
```

# c. Converting the DataTypes of numerical columns to numeric dataype

```python
# Converting the DataTypes of numerical columns to numeric dataype
raw_df['Dining Rating'].value_counts()
```

```python
# Replacing the text values with '0'
raw_df['Dining Rating'].replace(to_replace=['-','None'], value='0', inplace=True) 
```

```python
 # Checking for text values in the column before converting it to numeric datatype
raw_df['Dining Rating Count'].value_counts()
```

```python
# Replacing the text values with '0'
raw_df['Dining Rating Count'].replace(to_replace=['-','Does not offer Dining','Not enough Dining Reviews','None'], value='0', inplace=True)
```

```python
 raw_df.info()
```

```python
 # Changing Data Type of the numerical columns 
raw_df['Price for 2'] = raw_df['Price for 2'].astype('int64') 
```

```python
 raw_df.info()
```

```python
# Replacing the text values 'NA' with '0'
raw_df['Dining Rating'].replace(to_replace=['-','NA'], value='0', inplace=True) 
raw_df['Dining Rating Count'].replace(to_replace=['-','NA'], value='0', inplace=True) 
raw_df['Delivery Rating'].replace(to_replace=['-','NA'], value='0', inplace=True) 
raw_df['Delivery Rating Count'].replace(to_replace=['-','NA','Does not offer Delivery','Not enough Delivery Reviews'], value='0', inplace=True) 
```

```python
raw_df['Dining Rating'] = raw_df['Dining Rating'].astype('float64') 
raw_df['Dining Rating Count'] = raw_df['Dining Rating Count'].astype('int64')
raw_df['Delivery Rating'] = raw_df['Delivery Rating'].astype('float64')
raw_df['Delivery Rating Count'] = raw_df['Delivery Rating Count'].astype('int64')
```

```python
 raw_df.info()
```

```python
raw_df
```

# e. Removing the restaurant records whose Dining Rating, Dining Rating Count, Delivery Rating and Delivery Rating Count is 0

```python
 # Finding those restaurant whose has 0 Rating or Votes 
useless_data = (raw_df['Dining Rating'] == 0.0) & (raw_df['Dining Rating Count'] == 0) & (raw_df['Delivery Rating'] == 0.0) & (raw_df['Delivery Rating Count'] == 0)
```

```python
raw_df[useless_data]
```

```python
## Performing Negation of the useless dataset and then storing the correct data back in the raw_df DataFrame 
## This permamnently remove the wrong data from the original dataframe 
 
raw_df = raw_df[~useless_data] 
```

```python
raw_df
```

# Q10) Find the list of Affordable Restaurants?
# The criteria for Affordable Restaurants would be:
# 1) Low Price 2) High Rated
# First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
# Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
# restaurants that are 1/4th of 6000.

```python
max_price = zomato_df['Price for 2'].max() 
one_fourth_price = max_price/4 
one_fourth_price
```

```python
# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restaurant
aff_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features']] 
aff_rest_df = aff_rest_df[aff_rest_df['Price for 2'] <= 1250] 
aff_rest_df.sort_values(by='Price for 2', inplace=True) 
aff_rest_df
```

```python
 # Finding the highest rated list of restaurants
highrate_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features','Dining Rating']] 
highrate_rest_df = highrate_rest_df[highrate_rest_df['Dining Rating'] >= 4.5] 
highrate_rest_df.sort_values(by='Price for 2', inplace=True) 
highrate_rest_df
```

# Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !!

```python
 highrate_aff_df = pd.merge(aff_rest_df, highrate_rest_df, how='inner', on=['Name of Restaurant', 'Location']) 
highrate_aff_df = highrate_aff_df[['Name of Restaurant', 'Price for 2_x', 'Cuisine_x', 'Location', 'Features_x']] 
highrate_aff_df.rename(columns={'NAME':'Name of Restaurant', 'Price for 2_x':'Price for 2', 'Cuisine_x':'Cuisine',  
                                'REGION':'Location', 'Features_x':'Features'},inplace=True) 
```

```python
 # Affordable Restaurants with low price and high rating 
 highrate_aff_df 
```

# Q11) Find the list of most Reliable Restaurants?
# The criteria for most Reliable Restaurants would be:
# 1) Low Price 2) High Rated 3) Large No. of Votes
# First step will be to find the restaurants with Votes greater than Mean of Votes

```python
mean_votes = zomato_df['Dining Rating Count'].mean() 
mean_votes 
```

```python
# Finding list of restaurants that have Votes greater than and equal to  Mean of Vote 
mean_rest_df = zomato_df[['Name of Restaurant', 'Price for 2', 'Cuisine', 'Location', 'Features', 'Dining Rating Count']] 
mean_rest_df = mean_rest_df[mean_rest_df['Dining Rating Count'] > 177] 
mean_rest_df.sort_values(by='Dining Rating Count', inplace=True) 
mean_rest_df 
```

# These are the most reliable, highest rated and affordable restaurants:
# We obtain this dataframe by simply taking the intersection of highrate_aff_df & mean_rest_df
# This dataframe obtained below shows the restaurants whose:
# Cost is below 1250
# Rating is above 4.0
# Votes are above 107

```python
 reliable_rest_df = pd.merge(mean_rest_df, highrate_aff_df, how='inner', on=['Name of Restaurant', 'Location']) 
reliable_rest_df = reliable_rest_df[['Name of Restaurant', 'Price for 2_x', 'Cuisine_x', 'Location', 'Features_x']] 
reliable_rest_df.rename(columns={'Name of Restaurant':'Name of Restaurant', 'Price for 2_x':'Price for 2', 'Cuisine_x':'Cuisine',  
                                'Location':'Location', 'Features_x':'Features'},inplace=True) 
```
