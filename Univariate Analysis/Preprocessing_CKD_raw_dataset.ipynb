# Quan Qual function

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(dataset)
```

```python
dataset[quan]
```

```python
dataset[quan].isna().sum()
```

# On quan columns,we are replacing missing values with central tendency - Mean

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df_numericalData=imp.transform(dataset[quan])
```

```python
df_numericalData=pd.DataFrame(df_numericalData,columns=quan)
```

```python
df_numericalData
```

# On qual columns,we are replacing missing values with most_frequent values

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(dataset[qual])
df_categoricalData=imp.transform(dataset[qual])
df_categoricalData=pd.DataFrame(df_categoricalData,columns=qual)
```

```python
df_categoricalData
```

# Combining two preprocessed data frames

```python
Preprocessed_dataset=pd.merge(df_numericalData, df_categoricalData, left_index=True, right_index=True)
```
