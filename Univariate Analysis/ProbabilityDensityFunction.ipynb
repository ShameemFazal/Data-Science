# Probability Density Function

```python
def get_pdf_probability(dataset,startrange,endrange):
    # importing the required python libraries -- START 
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    # importing the required python libraries -- END

    
    # kde - kernel density function set as True to get the curve which set it as blue color and histogram in green color
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')

    
    # to draw red color vertical lines  -- START  
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # to draw red color vertical lines  -- END  
    
    # generate a sample
    sample=dataset

    # For distribution calculation, Mean and standard deviation are necessary
    #calculate parameters
    sample_mean=sample.mean()
    sample_std=sample.std()
    #print('Mean=%.3f','Standard Deviation=%.3f' % (sample_mean,sample_std))

    #define dsitribution
    dist=norm(sample_mean,sample_std)
    
    # sample probabilities for a range of outcomes
    # in-liner for-loop -> results will be converted into list in a single line
    values=[value for value in range(startrange,endrange)]
    
    # pdf is probability density function, 'values' will be passed to below forloop and probabilities are getting added
    probabilities=[dist.pdf(value) for value in values]
    
    # to view the whole probability, we take summation of all probabilities as shown below
    prob=sum(probabilities)
    
    print("The area between range({},{},{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["ssc_p"],60,70)
```

# Using normal for loop 

```python
def get_pdf_probability(dataset,startrange,endrange):
    # importing the required python libraries -- START 
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    # importing the required python libraries -- END

    
    # kde - kernel density function set as True to get the curve which set it as blue color and histogram in green color
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')

    
    # to draw red color vertical lines  -- START  
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # to draw red color vertical lines  -- END  
    
    # generate a sample
    sample=dataset

    # For distribution calculation, Mean and standard deviation are necessary
    #calculate parameters
    sample_mean=sample.mean()
    sample_std=sample.std()
    #print('Mean=%.3f','Standard Deviation=%.3f' % (sample_mean,sample_std))

    #define dsitribution
    dist=norm(sample_mean,sample_std)

    # Normal for-loop -- START
    values=[]
    probabilities=[]
    for value in  range(startrange,endrange):
        values.append(value)

    for value in values:
        probabilities.append(dist.pdf(value))
    # Normal for-loop -- END
    
    # to view the whole probability, we take summation of all probabilities as shown below
    prob=sum(probabilities)
    
    print("The area between range({},{},{}".format(startrange,endrange,sum(probabilities)))
    return prob
```

```python
get_pdf_probability(dataset["ssc_p"],50,70)
```
